```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
# Baca dataset
data = pd.read_csv('stock.csv')
```

```python
harga_pembelian = data['Purchase_Price']
print(harga_pembelian)
plt.figure(figsize=(10,5))
plt.hist(harga_pembelian, bins='auto',stacked=True, color='purple')
plt.ylabel('Frekuensi',fontsize = 10)
plt.title('1. Histogram dari Purchase Price',fontsize=15, color='black')
plt.xlim(xmin=min(harga_pembelian),xmax=max(harga_pembelian))
plt.show()
```

```python
#mengambil data top 5 berdasarkan Dividend_Share
top5 = (data.groupby('Symbol')['Dividend_Share']
                .sum()
                .reset_index()
                .sort_values(by='Dividend_Share',ascending=False)
                .head(5))
#Buat dataframe yang baru
top_5 = data[(data['Symbol'].isin(top5['Symbol'].to_list()))]
```

```python
plt.figure(figsize=(10,5))
plt.hist(top5.groupby('Symbol')['Dividend_Share'].nlargest(5), bins=10, stacked=True, range=(1,3), color='red')
plt.title('2. Histogram 5 Tertinggi Berdasarkan Dividend Share',fontsize=20, color='black')
plt.xlabel('Dividend Share', fontsize = 10)
plt.ylabel('Frekuensi',fontsize = 10)
plt.xlim(xmin=0.9,xmax=2.5)
```

```python
plt.figure(figsize=(10,5))
# mengelompokkan berdasarkan Symbol yang memiliki atribut Market_Value, sortir 5 tertinggi, dan mengurutkan dari terbesar ke terkecil
print(top_5.groupby('Symbol')['Market_Value'].nlargest(5).sort_values(ascending=False))
plt.hist(top_5.groupby('Symbol')['Market_Value'].nlargest(5), bins=10, stacked=True, range=(7000,15000), color='red')
plt.title('3. Histogram 5 Tertinggi Berdasarkan Market Value',fontsize=20, color='black')
plt.xlabel('Market Value', fontsize = 10)
plt.ylabel('Frekuensi',fontsize = 10)
plt.xlim(xmin=7000,xmax=15000)
```

```python
#Mengambil data top 7 berdasarkan Annual-Yield
top7 = (data.groupby('Symbol')['Annual_Yield']
                .sum()
                .reset_index()
                .sort_values(by='Annual_Yield',ascending=False)
                .head(7))
#Membuat dataframe baru
top_7 = data[(data['Symbol'].isin(top7['Symbol'].to_list()))]
# kelompokan berdasar Symbol yang memiliki Annual_Yield serta sortir top 7 tertinggi dan  urutkan dari terbesar ke terkecil
print(top7)
# plt.figure(figsize=(10,7))
plt.hist(top_7.groupby('Symbol')['Annual_Yield'].head(7).sort_values(ascending=True), bins=10, stacked=True, range=(0,6), color='red')
plt.title('4. Histogram 7 Tertinggi Berdasarkan Annual Yield',fontsize=20, color='black')
plt.xlabel('Annual Yield', fontsize = 10)
plt.ylabel('Frekuensi',fontsize = 10)
plt.xlim(xmin=0,xmax=6)
```
